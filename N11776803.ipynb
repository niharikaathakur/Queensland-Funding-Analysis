```python
import pandas as pd
url = "https://www.data.qld.gov.au/dataset/db190f2d-f866-4811-9a6e-4b78744b551b/resource/0f97b985-f5c7-49d2-8b0a-bc5dfbe070b9/download/advance-queensland-funding-recipients.csv"

#encoder
#reads
df = pd.read_csv(url,encoding = "latin1")

#view
df
#this code fetches data from an online CSV file, loads it into a pandas DataFrame
# and then displays the contents of the DataFrame.
```

```python
# Convert 'Approval date' to datetime
df['Approval date'] = pd.to_datetime(df['Approval date'],dayfirst=True, errors='coerce')
```

```python
# Convert 'Actual Contractual Commitment ($)' to numeric, errors are set to 'coerce' to handle any non-numeric values
df['Actual Contractual Commitment ($)'] = pd.to_numeric(df['Actual Contractual Commitment ($)'], errors='coerce')
```

```python
pd.options.mode.copy_on_write = True
# Exclude records with missing 'Actual Contractual Commitment ($)'
data_cleaned = df.dropna(subset=['Actual Contractual Commitment ($)'])

# Categorizing into South-East Queensland and Regional Queensland
# For simplicity, let's assume 'Brisbane and Redlands', 'Gold Coast', and 'Sunshine Coast' regions as South-East Queensland
# and the rest as Regional Queensland.

south_east_ql_regions = ['Brisbane and Redlands', 'Gold Coast', 'Sunshine Coast']
data_cleaned['Region Category'] = data_cleaned['RAP Region'].apply(lambda x: 'South-East Queensland' if x in south_east_ql_regions else 'Regional Queensland')

# Quick overview of distribution of entries between South-East Queensland and Regional Queensland
region_distribution = data_cleaned['Region Category'].value_counts()

region_distribution
```

1. Commenting on the geographical distribution

```python
import numpy as np


import plotly.express as px



# Extracting the regions and their counts
regions = region_distribution.index
counts = region_distribution.values

# Creating a DataFrame from the extracted data
data = {'Region Category': regions, 'Number of Entries': counts}
df = pd.DataFrame(data)

# Creating a bar plot using Plotly Express
fig = px.bar(df, x='Region Category', y='Number of Entries', color='Region Category',
             labels={'Region Category': 'Region Category', 'Number of Entries': 'Number of Entries'},
             title='Distribution of Entries between South-East Queensland and Regional Queensland')

# Displaying the plot
fig.show()
```

The analysis reveals a balanced distribution of funds between South-East Queensland and Regional Queensland. While South-East Queensland receives a larger portion of the funding due to its higher population and economic activity, a substantial amount of funding is also allocated to Regional Queensland to support innovation and development in rural and remote areas.

Question
2. How has the distribution of funds evolved over the years

```python


# Extracting year from 'Approval date'
data_cleaned['Year'] = data_cleaned['Approval date'].dt.year

# Creating a pivot table to prepare the data for Plotly Express
pivot_df = data_cleaned.pivot_table(index='Year', columns='Region Category', values='Actual Contractual Commitment ($)', aggfunc='sum')

# Resetting index to ensure 'Year' becomes a column
pivot_df.reset_index(inplace=True)

# Creating the bar plot using Plotly Express
fig = px.bar(pivot_df, x='Year', y=pivot_df.columns[1:], title='Advance Queensland Funding Trends Over Time',
             labels={'value': 'Total Funding Amount ($)', 'Year': 'Year'},
             category_orders={'Year': sorted(pivot_df['Year'].unique())})

fig.update_layout(xaxis={'type': 'category'}, xaxis_title='Year', yaxis_title='Total Funding Amount ($)',
                  legend_title='Region Category')

fig.show()



```

Funding allocations have varied over time, with peaks in certain years corresponding to significant government investments in innovation and economic development initiatives.

3. Top Funded Programs analysis

```python

# Grouping by program to see the distribution of funding
program_funding_distribution = data_cleaned.groupby('Program')['Actual Contractual Commitment ($)'].sum().sort_values(ascending=False)

# Selecting the top 10 programs by funding
top_programs = program_funding_distribution.head(10)

# Creating a DataFrame from the top programs
top_programs_df = top_programs.reset_index()

# Creating the bar plot using Plotly Express
fig = px.bar(top_programs_df, x='Actual Contractual Commitment ($)', y='Program',
             orientation='h', title='Top 10 Programs by Advance Queensland Funding',
             labels={'Actual Contractual Commitment ($)': 'Total Funding Amount ($)', 'Program': 'Program'})

fig.update_layout(xaxis_title='Total Funding Amount ($)', yaxis_title='Program',
                  yaxis={'categoryorder': 'total ascending'})

fig.show()

```

 This horizontal bar chart highlights the top-funded programs within the Advance Queensland initiative, indicating the areas of focus and strategic investment priorities.
